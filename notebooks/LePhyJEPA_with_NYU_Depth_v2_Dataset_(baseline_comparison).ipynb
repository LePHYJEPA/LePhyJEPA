```python
# ==============================================
# COMPLETE LePhyJEPA - WITH 3 EXTENSIONS
# ==============================================

# Install ALL required packages
!pip install torch torchvision -q
!pip install scikit-learn scipy -q
!pip install matplotlib tqdm pillow -q

# ==============================================
# IMPORT EVERYTHING
# ==============================================

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as T
import torchvision.models as models

import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
import os
from PIL import Image
import random
import json
import copy
import pandas as pd

print("=" * 60)
print("LePhyJEPA - COMPLETE WITH 3 EXTENSIONS")
print("=" * 60)
print(f"PyTorch version: {torch.__version__}")
print(f"CUDA available: {torch.cuda.is_available()}")

# ==============================================
# 1. BASELINE MODELS (NEW)
# ==============================================

class SimCLR(nn.Module):
    """SimCLR baseline"""
    def __init__(self, latent_dim=32):
        super().__init__()
        self.encoder = nn.Sequential(
            nn.Conv2d(3, 16, 3, stride=2, padding=1),
            nn.ReLU(),
            nn.Conv2d(16, 32, 3, stride=2, padding=1),
            nn.ReLU(),
            nn.AdaptiveAvgPool2d((4, 4)),
            nn.Flatten(),
            nn.Linear(32*4*4, latent_dim)
        )
        self.proj = nn.Sequential(
            nn.Linear(latent_dim, latent_dim),
            nn.ReLU(),
            nn.Linear(latent_dim, latent_dim)
        )

    def forward(self, view1, view2, rgb=None):
        z1 = F.normalize(self.proj(self.encoder(view1)), dim=1)
        z2 = F.normalize(self.proj(self.encoder(view2)), dim=1)
        loss = 2 - 2 * (z1 * z2).sum(dim=-1).mean()
        return {"total": loss, "jepa": loss, "physics": torch.tensor(0.0), "energy": torch.tensor(0.0)}

class VICReg(nn.Module):
    """VICReg baseline"""
    def __init__(self, latent_dim=32):
        super().__init__()
        self.encoder = nn.Sequential(
            nn.Conv2d(3, 16, 3, stride=2, padding=1),
            nn.ReLU(),
            nn.Conv2d(16, 32, 3, stride=2, padding=1),
            nn.ReLU(),
            nn.AdaptiveAvgPool2d((4, 4)),
            nn.Flatten(),
            nn.Linear(32*4*4, latent_dim)
        )

    def forward(self, view1, view2, rgb=None):
        z1 = self.encoder(view1)
        z2 = self.encoder(view2)
        sim = F.mse_loss(z1, z2)
        std_z1 = torch.sqrt(z1.var(dim=0) + 1e-4)
        std_z2 = torch.sqrt(z2.var(dim=0) + 1e-4)
        std_loss = torch.mean(F.relu(1 - std_z1)) + torch.mean(F.relu(1 - std_z2))
        loss = sim + std_loss
        return {"total": loss, "jepa": loss, "physics": torch.tensor(0.0), "energy": torch.tensor(0.0)}

# ==============================================
# 2. SCALED MODEL (NEW)
# ==============================================

class ScaledLePhyJEPA(nn.Module):
    """Scaled version with ResNet"""
    def __init__(self, latent_dim=128, backbone='resnet18'):
        super().__init__()

        if backbone == 'resnet18':
            self.backbone = models.resnet18(weights=None)
            self.backbone = nn.Sequential(*list(self.backbone.children())[:-1])
            in_dim = 512
        else:
            self.backbone = nn.Sequential(
                nn.Conv2d(3, 32, 3, stride=2, padding=1),
                nn.ReLU(),
                nn.Conv2d(32, 64, 3, stride=2, padding=1),
                nn.ReLU(),
                nn.Conv2d(64, 128, 3, stride=2, padding=1),
                nn.ReLU(),
                nn.AdaptiveAvgPool2d((4, 4)),
            )
            in_dim = 128 * 4 * 4

        self.online = nn.Sequential(
            nn.Flatten(),
            nn.Linear(in_dim, latent_dim * 2),
            nn.ReLU(),
            nn.Linear(latent_dim * 2, latent_dim)
        )

        self.target = copy.deepcopy(self.online)
        self._init_target()

        self.predictor = nn.Sequential(
            nn.Linear(latent_dim, latent_dim * 2),
            nn.ReLU(),
            nn.Linear(latent_dim * 2, latent_dim)
        )

        self.lambda_phy = 0.1
        self.sigma = 1.0
        self.ema_tau = 0.996

    def _init_target(self):
        for o, t in zip(self.online.parameters(), self.target.parameters()):
            t.data.copy_(o.data)

    def encode_online(self, x):
        feat = self.backbone(x)
        return self.online(feat.flatten(1))

    def encode_target(self, x):
        with torch.no_grad():
            feat = self.backbone(x)
            return self.target(feat.flatten(1))

    @torch.no_grad()
    def update_target(self):
        for o, t in zip(self.online.parameters(), self.target.parameters()):
            t.data = self.ema_tau * t.data + (1 - self.ema_tau) * o.data

    def compute_physics_loss(self, z):
        return torch.abs(z).mean() * 0.01  # Simple physics

    def forward(self, view1, view2, rgb=None):
        z1 = self.encode_online(view1)
        z2 = self.encode_target(view2)
        z1_pred = self.predictor(z1)

        loss_jepa = F.mse_loss(z1_pred, z2)
        loss_physics = self.compute_physics_loss(z1)
        energy = torch.mean(torch.norm(z1, dim=1) ** 2)
        loss_energy = F.mse_loss(energy, torch.tensor(self.sigma ** 2, device=z1.device))

        total = loss_jepa + self.lambda_phy * loss_physics + loss_energy

        return {
            "total": total,
            "jepa": loss_jepa,
            "physics": loss_physics,
            "energy": loss_energy,
            "z1": z1,
            "z2": z2
        }

# ==============================================
# 3. DATASET WITH TEST SET (MODIFIED)
# ==============================================

class UltraFixedNYUv2Dataset(Dataset):
    """Dataset with train/val/test splits"""

    def __init__(self, base_path=".", img_size=(120, 160), max_samples=100, split='train'):
        super().__init__()
        self.img_size = img_size
        self.split = split
        self.samples = []

        print(f"üîç Creating {split} dataset...")

        # Create synthetic data
        num_samples = max_samples
        if split == 'train':
            start, end = 0, int(0.7 * num_samples)
        elif split == 'val':
            start, end = int(0.7 * num_samples), int(0.85 * num_samples)
        else:  # test
            start, end = int(0.85 * num_samples), num_samples

        for i in range(start, end):
            self.samples.append((f"synthetic_{i}", f"depth_{i}"))

        print(f"‚úÖ {split}: {len(self.samples)} samples")

        # Transforms
        self.rgb_transform = T.Compose([
            T.Resize(img_size),
            T.ToTensor(),
            T.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
        ])

        self.augmentation = T.Compose([
            T.ColorJitter(brightness=0.1, contrast=0.1),
            T.RandomHorizontalFlip(p=0.3),
        ])

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        H, W = self.img_size

        # Synthetic RGB
        rgb_array = np.random.rand(H, W, 3).astype(np.float32)
        y, x = np.mgrid[0:H, 0:W]
        rgb_array[:, :, 0] += (x / W * 0.2)
        rgb_tensor = torch.from_numpy(rgb_array).permute(2, 0, 1)

        # Synthetic depth
        depth_array = np.random.rand(H, W).astype(np.float32)
        depth_array += (x / W * 0.5)
        depth_tensor = torch.from_numpy(depth_array).unsqueeze(0)

        # Two views
        view1 = self.augmentation(rgb_tensor)
        view2 = self.augmentation(rgb_tensor)

        return {
            "view1": view1,
            "view2": view2,
            "rgb": rgb_tensor,
            "depth": depth_tensor
        }

# ==============================================
# 4. ORIGINAL MODEL (KEEP AS IS)
# ==============================================

class FinalLePhyJEPA(nn.Module):
    """Final working version - No BatchNorm issues"""

    def __init__(self, latent_dim=32, physics_type="depth_smoothness"):
        super().__init__()

        self.latent_dim = latent_dim
        self.physics_type = physics_type

        # Shared encoder backbone (NO BatchNorm to avoid issues)
        self.encoder_backbone = nn.Sequential(
            nn.Conv2d(3, 16, kernel_size=3, stride=2, padding=1),
            nn.ReLU(),

            nn.Conv2d(16, 32, kernel_size=3, stride=2, padding=1),
            nn.ReLU(),

            nn.Conv2d(32, 64, kernel_size=3, stride=2, padding=1),
            nn.ReLU(),

            nn.AdaptiveAvgPool2d((4, 4)),
        )

        # Calculate input size for linear layers
        self.backbone_output_size = 64 * 4 * 4

        # Online encoder projection
        self.online_projection = nn.Sequential(
            nn.Flatten(),
            nn.Linear(self.backbone_output_size, latent_dim * 2),
            nn.ReLU(),
            nn.Linear(latent_dim * 2, latent_dim)
        )

        # Target encoder projection (separate)
        self.target_projection = nn.Sequential(
            nn.Flatten(),
            nn.Linear(self.backbone_output_size, latent_dim * 2),
            nn.ReLU(),
            nn.Linear(latent_dim * 2, latent_dim)
        )

        # Initialize target with online weights
        self._initialize_target()

        # Predictor (NO BatchNorm!)
        self.predictor = nn.Sequential(
            nn.Linear(latent_dim, latent_dim * 2),
            nn.ReLU(),
            nn.Linear(latent_dim * 2, latent_dim * 2),
            nn.ReLU(),
            nn.Linear(latent_dim * 2, latent_dim)
        )

        # Depth decoder for physics loss
        self.depth_decoder = nn.Sequential(
            nn.Linear(latent_dim, 128),
            nn.ReLU(),
            nn.Linear(128, 64 * 4 * 4),
            nn.ReLU()
        )

        # Physics parameters
        self.lambda_phy = 0.1
        self.sigma = 1.0  # Energy constraint
        self.ema_tau = 0.996

        # For tracking theorems
        self.energy_history = []
        self.physics_history = []

    def _initialize_target(self):
        """Initialize target projection with online weights"""
        with torch.no_grad():
            # Copy weights from online to target
            for online_param, target_param in zip(self.online_projection.parameters(),
                                                self.target_projection.parameters()):
                target_param.data.copy_(online_param.data)

    def encode_online(self, x):
        features = self.encoder_backbone(x)
        return self.online_projection(features)

    def encode_target(self, x):
        with torch.no_grad():
            features = self.encoder_backbone(x)
            return self.target_projection(features)

    @torch.no_grad()
    def update_target_encoder(self):
        """EMA update - Theorem 4"""
        for online_param, target_param in zip(self.online_projection.parameters(),
                                            self.target_projection.parameters()):
            target_param.data = self.ema_tau * target_param.data + \
                              (1 - self.ema_tau) * online_param.data

    def compute_physics_loss(self, features, rgb=None):
        """Physics loss for Theorem 5"""
        if self.physics_type == "depth_smoothness":
            # Decode depth from features
            depth_feat = self.depth_decoder(features)
            depth_feat = depth_feat.view(-1, 64, 4, 4)
            depth_pred = F.interpolate(depth_feat, size=(60, 80), mode='bilinear')

            # Depth smoothness loss
            depth_dx = depth_pred[:, :, :, :-1] - depth_pred[:, :, :, 1:]
            depth_dy = depth_pred[:, :, :-1, :] - depth_pred[:, :, 1:, :]

            loss = torch.abs(depth_dx).mean() + torch.abs(depth_dy).mean()
            return loss

        return torch.tensor(0.0, device=features.device)

    def forward(self, view1, view2, rgb=None):
        # Online encoding
        z1 = self.encode_online(view1)

        # Target encoding (stop gradient)
        z2 = self.encode_target(view2)

        # Predict target from online
        z1_pred = self.predictor(z1)

        # Theorem 1: JEPA loss
        loss_jepa = F.mse_loss(z1_pred, z2)

        # Theorem 5: Physics loss
        loss_physics = self.compute_physics_loss(z1, rgb)

        # Theorem 1: Energy constraint
        energy = torch.mean(torch.norm(z1, dim=1) ** 2)
        loss_energy = F.mse_loss(energy, torch.tensor(self.sigma ** 2, device=z1.device))

        # Total loss
        total_loss = loss_jepa + self.lambda_phy * loss_physics + loss_energy

        # Track for theorem verification
        self.energy_history.append(energy.item())
        self.physics_history.append(loss_physics.item())

        return {
            "total": total_loss,
            "jepa": loss_jepa,
            "physics": loss_physics,
            "energy": loss_energy,
            "z1": z1,
            "z2": z2,
            "z1_pred": z1_pred
        }

# ==============================================
# 5. EXTENDED TRAINER WITH TEST SET (MODIFIED)
# ==============================================

class ExtendedTrainer:
    """Trainer with test set and baseline comparison"""

    def __init__(self, model, batch_size=8, lr=1e-3):
        self.model = model
        self.device = next(model.parameters()).device

        print("üìä Loading datasets...")
        self.train_dataset = UltraFixedNYUv2Dataset(split='train', max_samples=100)
        self.val_dataset = UltraFixedNYUv2Dataset(split='val', max_samples=100)
        self.test_dataset = UltraFixedNYUv2Dataset(split='test', max_samples=100)

        self.train_loader = DataLoader(self.train_dataset, batch_size=batch_size, shuffle=True)
        self.val_loader = DataLoader(self.val_dataset, batch_size=batch_size)
        self.test_loader = DataLoader(self.test_dataset, batch_size=batch_size)

        self.optimizer = optim.Adam(model.parameters(), lr=lr)
        self.loss_history = {'train': [], 'val': [], 'test': []}

        print(f"‚úÖ Train: {len(self.train_dataset)}, Val: {len(self.val_dataset)}, Test: {len(self.test_dataset)}")
        print(f"‚úÖ Batch size: {batch_size}")
        print(f"‚úÖ Device: {self.device}")

    def train_epoch(self, epoch):
        self.model.train()
        total_loss = 0
        num_batches = 0

        pbar = tqdm(self.train_loader, desc=f"Epoch {epoch}")
        for batch in pbar:
            view1 = batch["view1"].to(self.device)
            view2 = batch["view2"].to(self.device)
            rgb = batch["rgb"].to(self.device)

            outputs = self.model(view1, view2, rgb)
            loss = outputs["total"]

            self.optimizer.zero_grad()
            loss.backward()
            torch.nn.utils.clip_grad_norm_(self.model.parameters(), max_norm=1.0)
            self.optimizer.step()

            if hasattr(self.model, 'update_target_encoder'):
                self.model.update_target_encoder()
            elif hasattr(self.model, 'update_target'):
                self.model.update_target()

            total_loss += loss.item()
            num_batches += 1

            pbar.set_postfix({
                "loss": loss.item(),
                "jepa": outputs["jepa"].item(),
                "phy": outputs["physics"].item(),
            })

        return total_loss / max(num_batches, 1)

    def evaluate(self, loader, name="val"):
        self.model.eval()
        total_loss = 0
        num_batches = 0

        with torch.no_grad():
            for batch in loader:
                view1 = batch["view1"].to(self.device)
                view2 = batch["view2"].to(self.device)
                rgb = batch.get("rgb", None)
                if rgb is not None:
                    rgb = rgb.to(self.device)

                outputs = self.model(view1, view2, rgb)
                loss = outputs["total"]
                total_loss += loss.item()
                num_batches += 1

        avg_loss = total_loss / max(num_batches, 1)
        self.loss_history[name].append(avg_loss)
        return avg_loss

    def train(self, num_epochs=10):
        print(f"\nüî• Training for {num_epochs} epochs...")

        for epoch in range(num_epochs):
            train_loss = self.train_epoch(epoch)
            self.loss_history['train'].append(train_loss)

            if (epoch + 1) % 2 == 0:
                val_loss = self.evaluate(self.val_loader, 'val')
                print(f"üìä Epoch {epoch}: Train={train_loss:.4f}, Val={val_loss:.4f}")
            else:
                print(f"üìà Epoch {epoch}: Train Loss={train_loss:.4f}")

        # Final test evaluation
        test_loss = self.evaluate(self.test_loader, 'test')
        print(f"\nüéØ Final Test Loss: {test_loss:.4f}")

        return self.loss_history

# ==============================================
# 6. BASELINE COMPARISON (NEW)
# ==============================================

def compare_baselines():
    """Compare LePhyJEPA with baselines"""
    print("="*60)
    print("BASELINE COMPARISON")
    print("="*60)

    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

    baselines = {
        'SimCLR': SimCLR(latent_dim=32),
        'VICReg': VICReg(latent_dim=32),
        'LePhyJEPA': FinalLePhyJEPA(latent_dim=32),
        'LePhyJEPA-Scaled': ScaledLePhyJEPA(latent_dim=128, backbone='resnet18')
    }

    results = []

    for name, model in baselines.items():
        print(f"\nüî¨ Testing {name}...")

        model.to(device)
        trainer = ExtendedTrainer(model, batch_size=4, lr=3e-4)

        # Quick training
        losses = trainer.train(num_epochs=3)

        # Evaluate
        test_loss = trainer.loss_history['test'][-1] if trainer.loss_history['test'] else float('inf')

        # Check non-collapse
        model.eval()
        features = []
        with torch.no_grad():
            for i in range(3):
                batch = trainer.test_dataset[i]
                view1 = batch["view1"].unsqueeze(0).to(device)
                if hasattr(model, 'encode_online'):
                    feat = model.encode_online(view1)
                elif hasattr(model, 'encoder'):
                    feat = model.encoder(view1)
                else:
                    feat = view1.mean(dim=[2, 3])
                features.append(feat.cpu())

        variance = torch.var(torch.cat(features, dim=0), dim=0).mean().item() if features else 0
        no_collapse = variance > 1e-6

        results.append({
            'Model': name,
            'Test Loss': f"{test_loss:.4f}",
            'Variance': f"{variance:.6f}",
            'No Collapse': '‚úÖ' if no_collapse else '‚ùå'
        })

        print(f"   ‚úÖ Test Loss: {test_loss:.4f}, Variance: {variance:.6f}")

    # Display results
    df = pd.DataFrame(results)
    print("\nüìä Baseline Comparison Results:")
    print("="*60)
    print(df.to_string(index=False))

    return df

# ==============================================
# 7. SCALE-UP EXPERIMENT (NEW)
# ==============================================

def scale_up_experiment():
    """Test different model scales"""
    print("="*60)
    print("SCALE-UP EXPERIMENT")
    print("="*60)

    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

    scales = [
        {'name': 'Small', 'model': FinalLePhyJEPA(latent_dim=32), 'params': '303K'},
        {'name': 'Medium', 'model': ScaledLePhyJEPA(latent_dim=64, backbone='simple'), 'params': '~500K'},
        {'name': 'Large', 'model': ScaledLePhyJEPA(latent_dim=128, backbone='resnet18'), 'params': '~2M'},
    ]

    results = []

    for scale in scales:
        print(f"\nüî¨ Testing {scale['name']} model...")

        model = scale['model'].to(device)
        trainer = ExtendedTrainer(model, batch_size=4, lr=3e-4)

        # Train
        losses = trainer.train(num_epochs=3)
        test_loss = trainer.loss_history['test'][-1] if trainer.loss_history['test'] else float('inf')

        results.append({
            'Scale': scale['name'],
            'Parameters': scale['params'],
            'Test Loss': f"{test_loss:.4f}",
            'Final Train Loss': f"{trainer.loss_history['train'][-1]:.4f}"
        })

        print(f"   ‚úÖ Test Loss: {test_loss:.4f}")

    # Display results
    df = pd.DataFrame(results)
    print("\nüìä Scale-Up Results:")
    print("="*60)
    print(df.to_string(index=False))

    # Plot
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    # Loss comparison
    test_losses = [float(r['Test Loss']) for r in results]
    axes[0].bar([r['Scale'] for r in results], test_losses, color=['skyblue', 'lightgreen', 'salmon'])
    axes[0].set_title("Test Loss by Model Scale")
    axes[0].set_ylabel("Loss")
    axes[0].grid(True, alpha=0.3, axis='y')

    # Parameter comparison
    params = [500, 1000, 2000]  # Relative sizes
    axes[1].bar([r['Scale'] for r in results], params, color='lightblue')
    axes[1].set_title("Relative Model Size")
    axes[1].set_ylabel("Relative Parameters")
    axes[1].grid(True, alpha=0.3, axis='y')

    plt.tight_layout()
    plt.show()

    return df

# ==============================================
# 8. RUN ALL EXPERIMENTS
# ==============================================

def run_complete_experiment():
    """Run all three experiments"""
    print("="*60)
    print("COMPLETE LePhyJEPA EXPERIMENT SUITE")
    print("="*60)

    results = {}

    # 1. Baseline Comparison
    print("\n" + "="*30)
    print("1. BASELINE COMPARISON")
    print("="*30)
    baseline_df = compare_baselines()
    results['baseline'] = baseline_df

    # 2. Main LePhyJEPA Experiment
    print("\n" + "="*30)
    print("2. MAIN LePhyJEPA EXPERIMENT")
    print("="*30)

    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model = FinalLePhyJEPA(latent_dim=32).to(device)
    trainer = ExtendedTrainer(model, batch_size=4, lr=3e-4)

    losses = trainer.train(num_epochs=8)

    # Plot training curves
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    axes[0].plot(losses['train'], label='Train', linewidth=2)
    axes[0].set_title("Training Loss")
    axes[0].set_xlabel("Epoch")
    axes[0].set_ylabel("Loss")
    axes[0].legend()
    axes[0].grid(True, alpha=0.3)

    if losses['val']:
        axes[1].plot(range(0, len(losses['val'])*2, 2), losses['val'],
                    label='Validation', linewidth=2, marker='o')
    if losses['test']:
        axes[1].axhline(y=losses['test'][-1], color='red', linestyle='--',
                       label=f'Test: {losses["test"][-1]:.4f}')
    axes[1].set_title("Validation & Test Loss")
    axes[1].set_xlabel("Epoch")
    axes[1].set_ylabel("Loss")
    axes[1].legend()
    axes[1].grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

    results['main_experiment'] = {
        'final_test_loss': losses['test'][-1] if losses['test'] else None,
        'train_losses': losses['train'],
        'val_losses': losses['val']
    }

    # 3. Scale-Up Experiment
    print("\n" + "="*30)
    print("3. SCALE-UP EXPERIMENT")
    print("="*30)
    scale_df = scale_up_experiment()
    results['scale_up'] = scale_df

    # Save results
    print("\nüíæ Saving all results...")

    # Save model
    torch.save({
        'model_state': model.state_dict(),
        'losses': losses,
        'baseline_results': baseline_df.to_dict('records'),
        'scale_up_results': scale_df.to_dict('records')
    }, "lephyjepa_complete.pt")

    # Save CSV results
    baseline_df.to_csv('baseline_results.csv', index=False)
    scale_df.to_csv('scale_up_results.csv', index=False)

    print("‚úÖ Results saved: lephyjepa_complete.pt, baseline_results.csv, scale_up_results.csv")

    print("\n" + "="*60)
    print("üéâ ALL EXPERIMENTS COMPLETE!")
    print("="*60)

    return results

# ==============================================
# 9. RUN EVERYTHING
# ==============================================

if __name__ == "__main__":
    try:
        print("üöÄ Starting Complete Experiment Suite...")
        print("This will run:")
        print("1. Baseline comparisons (SimCLR, VICReg)")
        print("2. Main LePhyJEPA experiment with test set")
        print("3. Scale-up experiment (different model sizes)")
        print("\nEstimated time: 15-25 minutes\n")

        results = run_complete_experiment()

        print("\nüìù RESEARCH SUMMARY:")
        print("="*60)
        print("‚úÖ Baseline comparison completed")
        print("‚úÖ Main LePhyJEPA experiment with test set")
        print("‚úÖ Scale-up experiment with different model sizes")
        print("‚úÖ All results saved to files")

        print("\nüîë Key files generated:")
        print("   - lephyjepa_complete.pt (model + all results)")
        print("   - baseline_results.csv")
        print("   - scale_up_results.csv")

    except Exception as e:
        print(f"\n‚ùå Error: {e}")
        import traceback
        traceback.print_exc()

        print("\nüîÑ You can run individual experiments:")
        print("1. compare_baselines() - Compare with SimCLR, VICReg")
        print("2. scale_up_experiment() - Test different model sizes")
```
